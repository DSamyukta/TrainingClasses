```python
import pandas as pd
```

```python
data = pd.read_csv('archive\iris.csv')
```

```python
data
```

## Separate Data based on Species

```python
setosa = data[data['Species'] == 'Iris-setosa']
setosa
```

```python
versicolor = data[data['Species'] == 'Iris-versicolor']
versicolor
```

```python
virginica = data[data['Species'] == 'Iris-virginica']
virginica
```

## Find mean of each feature/attribute

```python
print('Setosa - \n', setosa.mean())
print('Versicolor - \n', versicolor.mean())
print('Virginica - \n', virginica.mean())
```

## Find feature by which you can try to classify these Species
It will be nice to find one which has max S.D.; In our case it is PetalLengthCm

```python
setosa_upperlimit = 3
versicolor_upperlimit = 5
virginica_upperlimit = 6
```

```python
setosa_new = data[data['PetalLengthCm'] <= setosa_upperlimit]
setosa_new
```

```python
len(setosa_new) #Only one error
```

```python
x = data[data['PetalLengthCm'] <= versicolor_upperlimit] #Gives both setosa and versicolor
versicolor_new = x[x['PetalLengthCm'] >= setosa_upperlimit]
versicolor_new
```

```python
len(versicolor_new) #84 is missing and a few virginica is added, so considerable error 
```

```python
y = data[data['PetalLengthCm'] <= virginica_upperlimit] #Gives both setosa and versicolor
virginica_new = y[y['PetalLengthCm'] >= versicolor_upperlimit]
virginica_new
```

```python
len(virginica_new) #many errors here too
```

Hence, we are able to separate setosa with less error, other 2 have few or maybe many errors. So, its not classifying using one feature. Therefore, it is a multi-feature classification problem

There are other features and means too by which we can classify. Ex. mode, median, and etc.

```python
setosa.mode()
```

## Export the files

```python
setosa_new.to_excel('Setosa_new.xlsx')
versicolor_new.to_excel('Versicolor_new.xlsx')
virginica_new.to_excel('Virginica_new.xlsx')
```
