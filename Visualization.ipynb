```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

## Controlling line properties

### Using Keyword args

```python
plt.plot(t, t**2, linewidth=2.0)
```

```python
plt.plot(t, t**2, linewidth=20.0)
```

### Using setter methods

```python
line, = plt.plot(t, t**2, '^')
line.set_antialiased(False) # turn off antialising
```

### Using setp() command

```python
lines = plt.plot(t, t, t, t**2)
# use keyword args
plt.setp(lines, color='r', linewidth=2.0)
```

```python
# or MATLAB style string value pairs
plt.setp(lines, 'color', 'g', 'linewidth', 2.0)
```

```python
ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.1),
            )

plt.ylim(-2,2)
plt.show()
```

```python
ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2.05, 1.05), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.02),
            )

plt.ylim(-2,2)
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(221)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(222)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
import pandas as pd
```

```python
data = pd.read_csv('Iris.csv')
```

```python
data
```

```python
setosa = data[data['Species'] == 'Iris-setosa']
```

```python
setosa.describe()
```

```python
setosa.describe()['SepalLengthCm']['mean']
```

```python
import matplotlib.pyplot as plt
x1 = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']
y1 = [setosa.describe()['SepalLengthCm']['mean'], setosa.describe()['SepalWidthCm']['mean'], setosa.describe()['PetalLengthCm']['mean'], setosa.describe()['PetalWidthCm']['mean']]
plt.bar(x1,y1)
```

```python
versicolor = data[data['Species'] == 'Iris-versicolor']
```

```python
import numpy as np
SepalLength = np.mean(versicolor['SepalLengthCm'])
```

```python
SepalLength
```

```python
SepalWidth = np.mean(versicolor['SepalWidthCm'])
PetalLength = np.mean(versicolor['PetalLengthCm'])
PetalWidth = np.mean(versicolor['PetalWidthCm'])
```

```python
x2 = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']
y2 = [SepalLength, SepalWidth, PetalLength, PetalWidth]
plt.bar(x2,y2)
```

```python
virginica = data[data['Species'] == 'Iris-virginica']
```

```python
SepalLength3 = np.mean(virginica['SepalLengthCm'])
SepalWidth3 = np.mean(virginica['SepalWidthCm'])
PetalLength3 = np.mean(virginica['PetalLengthCm'])
PetalWidth3 = np.mean(virginica['PetalWidthCm'])
```

```python
x3 = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']
y3 = [SepalLength3, SepalWidth3, PetalLength3, PetalWidth3]
plt.bar(x3,y3)
```

```python
x_axis = np.arange(len(x1))
plt.bar(x_axis-0.2, y1,0.2, label = 'setosa')
plt.bar(x_axis, y2,0.2, label = 'versicolor')
plt.bar(x_axis+0.2, y3,0.2, label = 'virginica')

plt.xticks(x_axis,x1)
plt.show()
```
